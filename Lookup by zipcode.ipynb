```python
import pandas as pd

load_data = lambda x, **kwargs: pd.read_csv(x, sep='\t', **kwargs)

zipcode_table = load_data('UCMR3_ZipCodes.txt')

zipcode_to_PWSID = lambda x: list(zipcode_table[zipcode_table['ZIPCODE'] == x]['PWSID'])[0]

results = load_data('UCMR3_All.txt', encoding='latin1')

#drt_table = load_data('UCMR3_DRT.txt')
```

```python
#list(zipcode_table[zipcode_table['ZIPCODE'] == '93728']['PWSID'])[0]
our_pwsid = zipcode_to_PWSID('93728')
our_pwsid
```

```python
f = results['PWSID'] == our_pwsid
u = results['AnalyticalResultsSign'] == '='
our_rows = results[f&u]
our_rows
```

```python

ref_levels = pd.read_csv('epa_regulation_levels.csv', index_col='Contaminant')
```

```python
ref_levels.ix['cobalt', 'Reference Concentration']
```

```python
doselect = lambda x: ref_levels.ix[our_rows.ix[x, 'Contaminant'], 'Reference Concentration'] <= our_rows.ix[x, 'AnalyticalResultValue'] 
exceeded_levels = our_rows.select(doselect)
```
